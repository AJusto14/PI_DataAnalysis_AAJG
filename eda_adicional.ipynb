Ahora veamos una gráfica que mostrará por provincia los accesos que se han hecho a lo largo de los años.

```python
#Primero separamos los datos que usaremos
anio_x_provincia = provincia.groupby(['Provincia', 'Año'])['Accesos telefonía fija (total)'].sum().reset_index()
```

```python
#Configuramos la gráfica
plt.figure(figsize=(12, 6))
for province in anio_x_provincia['Provincia'].unique():
    province_data = anio_x_provincia[anio_x_provincia['Provincia'] == province]
    plt.plot(province_data['Año'], province_data['Accesos telefonía fija (total)'], label=province)

plt.xlabel('Año')
plt.ylabel('Accesos telefonía fija (total)')
plt.title('Suma de Accesos telefonía fija (total) por Provincia y Año')
plt.legend(loc='upper left')
plt.show()
```

Vemos como siempre que la capital es quien tiene los accesos de la telefonía con más movimiento, pero ha ido disminuyendo a lo largo de los años.
